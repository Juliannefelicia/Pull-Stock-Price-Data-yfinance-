```python
import datetime
import yfinance as yf
import pandas as pd

pd.set_option('display.max_columns',None)
pd.set_option('display.width',1000)
def getStock(stkInput, dayInput):
    dt = datetime.date.today()
    dtPast = dt + datetime.timedelta(days=-dayInput)

    # Call Yahoo finance to get stock data for the stock provided.
    dfStock = yf.download(stkInput, start=dtPast, end=dt)
    return dfStock

def removeTime(df):
    newDateList=[]
    dateList = list(df.index.values)
    print(dateList)
    for i in range(0, len(dateList)):
        dateStr=str(dateList[i])
        itemList=dateStr.split("T")
        print(itemList)
        newDateList.append(itemList[0])
    df["Date"] = newDateList
    print(df)
    df.set_index("Date", inplace=True)

while True:
    print("-------------------------------------------------")
    print("Stock Report Menu Options")
    print("-------------------------------------------------")
    optionValue=int(input("\n1. Report changes for a stock"
                          "\n2. Quit"
                          "\n"))
    if optionValue==1:
        stockSymbol = input("Please enter the stock symbol: ")
        days = int(input("Please enter the number of days for the analysis: "))
        df = getStock(stockSymbol, days)
        dt = datetime.date.today()
        dtPast = dt + datetime.timedelta(days=-days)
        df2=df[["Close","Volume"]].copy()
        volChg = []
        closeChg = []
        for i in range(len(df)):
            if i==0:
                volChg.append(0.0000)
                closeChg.append(0.0000)
            else:
                # df2.iloc[0, 0] #Date is index, not column
                closeChg.append(round((df2.iloc[i,0]-df2.iloc[i-1,0])/df2.iloc[i-1,0],4))
                volChg.append(round((df2.iloc[i, 1] - df2.iloc[i - 1, 1]) / df2.iloc[i - 1, 1],4))
        df2['Volume % Change']=volChg
        df2['Close % Change'] = closeChg


#The formula for daily percent change = (Current day close price−Previous day close price/Previous day close price)
#The formula for the overall percentage change = (Period end close price−Period start close price/Period start close price)

        print("************************************************************")
        print("Daily Percent Changes - " + str(dtPast) + " to " + str(dt) + " * " + str(stockSymbol) + " *")
        print("************************************************************\n", df2)


# Close, Volume, Volume % Change, Close % Change,
# % Volume Change, % Close Price Change

        print("\n------------------------------------------------------------")
        print("Summary of Cumulative Changes for " + str(stockSymbol))
        print("------------------------------------------------------------")
        print(str(dtPast) + " to " + str(dt))
        print()
        periodCloseChg=((df2.iloc[-1, 0] - df2.iloc[0, 0]) / df2.iloc[0, 0])
        periodVolChg=((df2.iloc[-1, 1] - df2.iloc[0, 1]) / df2.iloc[0, 1])
        print("% Volume Change: "+str(round(periodVolChg,3)))
        print("% Close Price Change: "+str(round(periodCloseChg,3)))
    elif optionValue==2:
        print("Goodbye")
        break
    else:
        print("Wrong option, try again")
```
